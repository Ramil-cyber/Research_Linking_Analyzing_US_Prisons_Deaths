```python
import pandas as pd

BJS_prison = pd.read_csv(
    "https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/mciprison2016_20192025.csv"
)
BJS_prison
```

```python
# ------------------------------------------------------------------
# Keeping only the specified columns
# ------------------------------------------------------------------
cols_to_keep = [
    "State",
    "Year",
    "Lname",
    "Fname",
    "Dobyear",
    "Facname",
    "Facloc",
    "Gender",
    "Race",
    "Cause",
]
BJS_prison_reduced = BJS_prison[cols_to_keep].copy()

# ------------------------------------------------------------------
# Renaming columns as requested
# ------------------------------------------------------------------
rename_map = {
    "State": "state_name",
    "Year": "death_year",
    "Lname": "last_name",
    "Fname": "first_name",
    "Dobyear": "birth_year",
    "Facname": "facility_name",
    "Facloc": "city_name",
    "Cause": "death_cause",
}
BJS_prison_reduced.rename(columns=rename_map, inplace=True)

# ------------------------------------------------------------------
# Maping numeric codes to descriptive labels
# ------------------------------------------------------------------

state_map = {
    1: "Alabama",
    2: "Alaska",
    3: "Arizona",
    4: "Arkansas",
    5: "California",
    6: "Colorado",
    7: "Connecticut",
    8: "Delaware",
    10: "Florida",
    11: "Georgia",
    12: "Hawaii",
    13: "Idaho",
    14: "Illinois",
    15: "Indiana",
    16: "Iowa",
    17: "Kansas",
    18: "Kentucky",
    19: "Louisiana",
    20: "Maine",
    21: "Maryland",
    22: "Massachusetts",
    23: "Michigan",
    24: "Minnesota",
    25: "Mississippi",
    26: "Missouri",
    27: "Montana",
    28: "Nebraska",
    29: "Nevada",
    30: "New Hampshire",
    31: "New Jersey",
    32: "New Mexico",
    33: "New York",
    34: "North Carolina",
    35: "North Dakota",
    36: "Ohio",
    37: "Oklahoma",
    38: "Oregon",
    39: "Pennsylvania",
    40: "Rhode Island",
    41: "South Carolina",
    42: "South Dakota",
    43: "Tennessee",
    44: "Texas",
    45: "Utah",
    46: "Vermont",
    47: "Virginia",
    48: "Washington",
    49: "West Virginia",
    50: "Wisconsin",
    51: "Wyoming",
}
BJS_prison_reduced["state_name"] = pd.to_numeric(
    BJS_prison_reduced["state_name"], errors="coerce"
).map(state_map)

BJS_prison_reduced["birth_year"] = pd.to_numeric(
    BJS_prison_reduced["birth_year"], errors="coerce"
).astype("Int64")

gender_map = {1: "Male", 2: "Female"}
BJS_prison_reduced["Gender"] = (
    BJS_prison_reduced["Gender"].map(gender_map).fillna("Missing/Not Provided")
)

race_map = {
    1: "White",
    2: "Black",
    3: "Hispanic / Latino",
    4: "American Indian / Alaska Native",
    5: "Asian",
    6: "Native Hawaiian / Other Pacific Islander",
    7: "Two or more races",
    8: "Additional racial category",
    9: "Not Known",
}
BJS_prison_reduced["Race"] = (
    BJS_prison_reduced["Race"].map(race_map).fillna("Missing/Not Provided")
)

cause_map = {
    1: "Illness",
    2: "Illness (AIDS-related)",
    3: "Alcohol/drug intoxication",
    4: "Accidental injury to self",
    5: "Accidental injury by other",
    6: "Suicide",
    7: "Homicide",
    8: "Other",
    11: "Homicide",
}
BJS_prison_reduced["death_cause"] = (
    pd.to_numeric(BJS_prison_reduced["death_cause"], errors="coerce")
    .map(cause_map)
    .fillna("Missing/Not Provided")
)

# ------------------------------------------------------------------
# Uppercasing all string (object) columns
# ------------------------------------------------------------------
for col in BJS_prison_reduced.select_dtypes(include=["object"]).columns:
    BJS_prison_reduced[col] = BJS_prison_reduced[col].str.upper()

# ------------------------------------------------------------------
# Renaming 'Gender' and 'Race' columns to lowercase
# ------------------------------------------------------------------
BJS_prison_reduced.rename(columns={"Gender": "gender", "Race": "race"}, inplace=True)

# ------------------------------------------------------------------
# Reordering columns as specified
# ------------------------------------------------------------------
desired_order = [
    "state_name",
    "city_name",
    "facility_name",
    "first_name",
    "last_name",
    "birth_year",
    "death_year",
    "gender",
    "race",
    "death_cause",
]
BJS_prison_reduced = BJS_prison_reduced[desired_order]
```

```python
# ------------------------------------------------------------------
# Creating 'age' column
# ------------------------------------------------------------------
BJS_prison_reduced["age"] = (
    BJS_prison_reduced["death_year"] - BJS_prison_reduced["birth_year"]
)

# ------------------------------------------------------------------
# Inserting 'age' column between 'birth_year' and 'death_year'
# ------------------------------------------------------------------
cols = BJS_prison_reduced.columns.tolist()
# Remove 'age' from end
cols.remove("age")
# Find index of 'birth_year'
idx = cols.index("birth_year")
# Insert 'age' right after
cols.insert(idx + 1, "age")
# Reorder DataFrame
BJS_prison_reduced = BJS_prison_reduced[cols]
```

```python
# Computing missing counts and percentages
missing_counts = BJS_prison_reduced.isna().sum()
missing_percent = (BJS_prison_reduced.isna().mean() * 100).round(2)

# Combining into summary DataFrame
missing_summary = pd.DataFrame(
    {"missing_count": missing_counts, "missing_percent": missing_percent}
)
missing_summary
```

```python
# Defining age bins and labels for 'age_range'
# ------------------------------------------------------------------
bins = [-1, 20, 30, 40, 50, 60, 70, float("inf")]
labels = [" 20", "21-30", "31-40", "41-50", "51-60", "61-70", " 71"]

# ------------------------------------------------------------------
# Creating 'age_range' using pd.cut
# ------------------------------------------------------------------
BJS_prison_reduced["age_range"] = pd.cut(
    BJS_prison_reduced["age"], bins=bins, labels=labels, right=True, include_lowest=True
)

# ------------------------------------------------------------------
# Inserting 'age_range' immediately after 'age'
# ------------------------------------------------------------------
cols = BJS_prison_reduced.columns.tolist()
cols.remove("age_range")  # remove from end
age_idx = cols.index("age")  # find 'age'
cols.insert(age_idx + 1, "age_range")  # insert after 'age'
BJS_prison_reduced = BJS_prison_reduced[cols]

# ------------------------------------------------------------------
# Saving the final cleaned & mapped DataFrame
# ------------------------------------------------------------------
output_path = "BJS_prison_cleaned_v2.csv"
BJS_prison_reduced.to_csv(output_path, index=False)

print(f"Saved mapped dataset to: {output_path}")
BJS_prison_reduced.sample(10)
```

```python
# ------------------------------------------------------------------
# Grouping by 'state_name' and 'death_year' and count records
# ------------------------------------------------------------------
BJS_prison_grouped = (
    BJS_prison_reduced.groupby(
        ["state_name", "death_year"], as_index=False
    )
    .size()
    .rename(columns={"size": "death_count"})
)

# ------------------------------------------------------------------
# Saving the grouped summary to CSV
# ------------------------------------------------------------------
output_path = "BJS_prison_grouped_by_state_year.csv"
BJS_prison_grouped.to_csv(output_path, index=False)

print(f"Grouped summary saved to: {output_path}")
BJS_prison_grouped.sample(10)
```

```python
import pandas as pd
                                    
Prison_pop = pd.read_csv(
    "https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/Prison_population_2015_2022_updated.csv"
)

# ------------------------------------------------------------------
# Uppercasing all string (object) columns
# ------------------------------------------------------------------
for col in Prison_pop.select_dtypes(include=["object"]).columns:
    Prison_pop[col] = Prison_pop[col].str.upper()

Prison_pop.sample(10)
```

```python
# ------------------------------------------------------------------
# Defining columns to keep
# ------------------------------------------------------------------
keep_cols = [
    'year',
    'state_abbr',
    'state_name',
    'total_pop_15to64',
    'female_pop_15to64',
    'male_pop_15to64',
    'aapi_pct',
    'black_pct',
    'latinx_pct',
    'native_pct',
    'white_pct',
    'total_incarceration',
    'total_incarceration_rate',
    'total_prison_pop',
    'female_prison_pop',
    'male_prison_pop'
]

# ------------------------------------------------------------------
# Subseting to those columns
# ------------------------------------------------------------------
Prison_pop_reduced = Prison_pop[keep_cols]

# ------------------------------------------------------------------
# Filtering to years 20152019 (inclusive)
# ------------------------------------------------------------------
Prison_pop_filtered = Prison_pop_reduced[Prison_pop_reduced['year'].between(2015, 2019)]

# Renaming 'year' to 'death_year'
Prison_pop_filtered = Prison_pop_filtered.rename(columns={'year': 'death_year'})


# ------------------------------------------------------------------
# Saving the reduced dataset
# ------------------------------------------------------------------
output_path = 'Prison_pop_filtered.csv'
Prison_pop_filtered.to_csv(output_path, index=False)

print(f"Saved reduced dataset with {len(keep_cols)} columns to:\n{output_path}")
print("Columns retained:")
print(Prison_pop_filtered.columns.tolist())
print("\nPreview:")
Prison_pop_filtered.sample(10)
```

```python
# ------------------------------------------------------------------
# Identifing and display common columns
# ------------------------------------------------------------------
common_cols = list(set(BJS_prison_grouped.columns).intersection(Prison_pop_filtered.columns))
print("Common columns:", common_cols)

# ------------------------------------------------------------------
# Merging on death_year and state_name, preserving all other fields
# ------------------------------------------------------------------
Merged_full_df = pd.merge(
    BJS_prison_grouped,
    Prison_pop_filtered,
    on=['death_year', 'state_name'],
    how='inner',  # use 'outer' if you want to keep unmatched rows
    suffixes=('_1', '_2')
)

# ------------------------------------------------------------------
# Saving the merged dataset
# ------------------------------------------------------------------
output_path = 'Last_merged_full_df.csv'
Merged_full_df.to_csv(output_path, index=False)


print(f"Merged file saved to: {output_path}")
Merged_full_df.sample(10)
```

```python
# Computing missing counts and percentages
missing_counts = Merged_full_df.isna().sum()
missing_percent = (Merged_full_df.isna().mean() * 100).round(2)

# Combining into summary DataFrame
missing_summary = pd.DataFrame(
    {"missing_count": missing_counts, "missing_percent": missing_percent}
)
missing_summary
```
