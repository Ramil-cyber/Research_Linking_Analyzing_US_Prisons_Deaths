```python
import pandas as pd

BJA = pd.read_csv(
    "https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/BJA_updated_fields_Full-Data_data-882_Cleaned-up.2.25.csv"
)
BJA
```

```python
counts_BJA = BJA.isnull().sum()
counts_BJA
```

```python
# Renaming variables for merging
rename_dict = {
    "Manner of Death": "Death_Cause",
    "Birth Year": "Birth_Year",
    "Calendar Year Death": "Death_Year",
    "Decedent Last Name": "Last_Name",
    "First Name": "First_Name",
    "Grantee Legal Name": "Facility_name",
    "Year of Fiscal Year of Death": "Reporting_Death_Year",
}

# Specifying columns to keep original names
keep_columns = list(rename_dict.keys()) + ["Gender", "Race", "State", "Age", "City"]

# Subseting and renaming
BJA_clean = BJA[keep_columns].rename(columns=rename_dict)

# Adding new column 'Source'
BJA_clean["Source"] = "BJA"

# Reseting index to remove the numbered index column
BJA_clean.reset_index(drop=True, inplace=True)
BJA_clean
```

```python
# Droping any auto-generated index columns
BJA_clean = BJA_clean.loc[:, ~BJA_clean.columns.str.match(r"Unnamed")]

# Coercing Birth_Year to numeric, clear out 9999 placeholders, then convert to Int64
BJA_clean["Birth_Year"] = pd.to_numeric(BJA_clean["Birth_Year"], errors="coerce")
BJA_clean.loc[BJA_clean["Birth_Year"] == 9999, "Birth_Year"] = pd.NA
BJA_clean["Birth_Year"] = BJA_clean["Birth_Year"].astype("Int64")

# Coercing Age to numeric, replace 0s with NA, fill missing via Death_Year – Birth_Year, then convert to Int64
BJA_clean["Age"] = pd.to_numeric(BJA_clean["Age"], errors="coerce")
BJA_clean.loc[BJA_clean["Age"] == 0, "Age"] = pd.NA
mask_age_na = (
    BJA_clean["Age"].isna()
    & BJA_clean["Death_Year"].notna()
    & BJA_clean["Birth_Year"].notna()
)
BJA_clean.loc[mask_age_na, "Age"] = (
    BJA_clean.loc[mask_age_na, "Death_Year"] - BJA_clean.loc[mask_age_na, "Birth_Year"]
)
BJA_clean["Age"] = BJA_clean["Age"].astype("Int64")

# Renaming columns as requested
BJA_clean = BJA_clean.rename(
    columns={
        "State": "State_Name",
        "City": "City_Name",
        "Facility_name": "Facility_Name",
    }
)

# Reorderoing columns in the desired sequence
cols_order = [
    "First_Name",
    "Last_Name",
    "Gender",
    "Race",
    "Birth_Year",
    "Death_Year",
    "Reporting_Death_Year",
    "Age",
    "State_Name",
    "City_Name",
    "Facility_Name",
    "Death_Cause",
    "Source",
]
BJA_clean = BJA_clean[cols_order]

```

```python
print(BJA_clean['Race'].dropna().unique())
```

```python
# Defining the replacement mapping for Race
race_replace = {
    'White': 'White (not Hispanic)',
    'Black': 'Black or African American',
    'Unknown': 'Not Known',
    'American Indian or Alaska Native': 'American Indian/Alaska Native',
    'Asian': 'Asian (not Hispanic)',
    'Multi': 'Two or more races (not Hispanic)'
}

# 3) Apply the mapping to the Race column
BJA_clean['Race'] = BJA_clean['Race'].replace(race_replace)

# 2) Define mappings for Death_Cause replacements
death_cause_map = {
    'Homicide (e.g., an incident between two or more incarcerated individuals resulting in a death)': 'Homicide',
    'Death attributed to use of force by a law enforcement or corrections officer': 'Use of Force',
    'Natural causes': 'Illness'
}

# 3) Apply the replacements
BJA_clean['Death_Cause'] = BJA_clean['Death_Cause'].replace(death_cause_map)

# 4) (Optional) Verify the update
print(BJA_clean['Race'].value_counts())

# # 5) Save the updated dataset
# output_path = "BJA_cleaned.csv"
# BJA_clean.to_csv(output_path, index=False)
# print(f"Updated dataset saved to: {output_path}")
```

```python
# 2) Capitalize Gender values
# BJA_clean['Gender'] = BJA_clean['Gender'].map({'male': 'Male', 'female': 'Female'})

# 3) Drop the Death_Year and Age columns
BJA_clean = BJA_clean.drop(columns=['Death_Year', 'Age'])

# 4) Reorder columns as requested
cols_order = [
    'First_Name',
    'Last_Name',
    'Gender',
    'Race',
    'Birth_Year',
    'Reporting_Death_Year',
    'State_Name',
    'City_Name',
    'Facility_Name',
    'Death_Cause',
    'Source'
]
BJA_clean = BJA_clean[cols_order]

# 2) Map state abbreviations to full state names
abbr_to_full = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas',
    'CA': 'California', 'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware',
    'DC': 'District of Columbia', 'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii',
    'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada',
    'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York',
    'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah',
    'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming'
}

BJA_clean['State_Name'] = BJA_clean['State_Name'].map(abbr_to_full)


```

```python
# 2) Sort by State_Name alphabetically
BJA_clean = BJA_clean.sort_values(by='State_Name')

# 3) Reset index (optional)
BJA_clean = BJA_clean.reset_index(drop=True)

# 5) Save the updated dataset
output_path = "BJA_cleaned.csv"
BJA_clean.to_csv(output_path, index=False)
print(f"Updated dataset saved to: {output_path}")
```
