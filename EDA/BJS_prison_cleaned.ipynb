```python
import pandas as pd

BJS_prison = pd.read_csv(
    "https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/mciprison2016_20192025.csv"
)
BJS_prison
```

```python
# BJS_prison.info()
counts_BJS = BJS_prison.isnull().sum()
counts_BJS
```

```python
# 2) Rename columns
rename_map = {
    'State': 'State_Name',
    'Lname': 'Last_Name',
    'Fname': 'First_Name',
    'Dobyear': 'Birth_Year',
    'Facname': 'Facility_Name',
    'Year': 'Reporting_Death_Year',
    'Cause': 'Death_Cause',
    'Facloc': 'City_Name'
}
BJS_prison_updated = BJS_prison.rename(columns=rename_map)

# 3) Add Source column
BJS_prison_updated['Source'] = 'BJS'

# 4) Convert Reporting_Death_Year and Birth_Year to integer
BJS_prison_updated['Reporting_Death_Year'] = pd.to_numeric(
    BJS_prison_updated['Reporting_Death_Year'], errors='coerce'
).astype('Int64')
BJS_prison_updated['Birth_Year'] = pd.to_numeric(
    BJS_prison_updated['Birth_Year'], errors='coerce'
).astype('Int64')

# 5) Create Death_Year (duplicate of Reporting_Death_Year) and Age
BJS_prison_updated['Death_Year'] = BJS_prison_updated['Reporting_Death_Year']
BJS_prison_updated['Age'] = (
    BJS_prison_updated['Death_Year'] - BJS_prison_updated['Birth_Year']
).astype('Int64')

# 6) Map State codes to USPS abbreviations
state_map = {
    1:'AL', 2:'AK', 3:'AZ', 4:'AR', 5:'CA', 6:'CO', 7:'CT', 8:'DE', 9:'DC',
    10:'FL',11:'GA',12:'HI',13:'ID',14:'IL',15:'IN',16:'IA',17:'KS',18:'KY',19:'LA',
    20:'ME',21:'MD',22:'MA',23:'MI',24:'MN',25:'MS',26:'MO',27:'MT',28:'NE',29:'NV',
    30:'NH',31:'NJ',32:'NM',33:'NY',34:'NC',35:'ND',36:'OH',37:'OK',38:'OR',39:'PA',
    40:'RI',41:'SC',42:'SD',43:'TN',44:'TX',45:'UT',46:'VT',47:'VA',48:'WA',49:'WV',
    50:'WI',51:'WY'
}
BJS_prison_updated['State_Name'] = pd.to_numeric(
    BJS_prison_updated['State_Name'], errors='coerce'
).map(state_map)

# 7) Map Gender and Race codes to labels
BJS_prison_updated['Gender'] = BJS_prison_updated['Gender'].map({1:'male', 2:'female'})
race_map = {
    1:'White (not Hispanic)', 2:'Black or African American', 3:'Hispanic or Latino',
    4:'American Indian/Alaska Native', 5:'Asian (not Hispanic)',
    6:'Native Hawaiian or Other Pacific Islander',
    7:'Two or more races (not Hispanic)', 8:'Additional racial categories', 9:'Not Known'
}
BJS_prison_updated['Race'] = BJS_prison_updated['Race'].map(race_map)

# 8) Keep only the specified columns
keep_cols = [
    'State_Name', 'Last_Name', 'First_Name', 'Birth_Year',
    'Facility_Name', 'Reporting_Death_Year', 'Death_Cause',
    'City_Name', 'Gender', 'Race', 'Death_Year', 'Source', 'Age'
]
BJS_prison_updated = BJS_prison_updated[keep_cols]

# 9) Convert Death_Cause to nullable integer, blank out Death_Year and Age
BJS_prison_updated['Death_Cause'] = pd.to_numeric(BJS_prison_updated['Death_Cause'], errors='coerce').astype('Int64')
BJS_prison_updated['Death_Year'] = pd.NA
BJS_prison_updated['Age'] = pd.NA

# # 10) Save the resulting dataset
# output_path = 'EDA.csv'
# BJS_prison_updated.to_csv(output_path, index=False)
# print(f"Final cleaned dataset saved to: {output_path}")
```

```python
# 2) Ensure Death_Cause is numeric
BJS_prison_updated['Death_Cause'] = pd.to_numeric(BJS_prison_updated['Death_Cause'], errors='coerce')

# 3) Map codes to simplified categories
death_cause_map = {
    1:  'Illness',
    2:  'Illness',
    3:  'Accident',
    4:  'Accident',
    5:  'Accident',
    6:  'Suicide',
    7:  'Homicide',
    8:  'Other',
    11: 'Homicide'
}

BJS_prison_updated['Death_Cause'] = BJS_prison_updated['Death_Cause'].map(death_cause_map)

# 4) (Optional) Quick check
print(BJS_prison_updated['Death_Cause'].value_counts(dropna=False))

```

```python
# 2) Capitalize Gender values
BJS_prison_updated['Gender'] = BJS_prison_updated['Gender'].map({'male': 'Male', 'female': 'Female'})

# 3) Drop unwanted columns
BJS_prison_updated = BJS_prison_updated.drop(columns=['Death_Year', 'Age'])

# 4) Reorder columns as specified
cols_order = [
    'First_Name', 'Last_Name', 'Gender', 'Race',
    'Birth_Year', 'Reporting_Death_Year',
    'State_Name', 'City_Name', 'Facility_Name',
    'Death_Cause', 'Source'
]
BJS_prison_updated = BJS_prison_updated[cols_order]

# 2) Map state abbreviations to full state names
abbr_to_full = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas',
    'CA': 'California', 'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware',
    'DC': 'District of Columbia', 'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii',
    'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada',
    'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York',
    'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah',
    'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming'
}

BJS_prison_updated['State_Name'] = BJS_prison_updated['State_Name'].map(abbr_to_full)

# 5) Save the final dataset
output_path = 'BJS_cleaned.csv'
BJS_prison_updated.to_csv(output_path, index=False)
print(f"Updated dataset saved to: {output_path}")
```
